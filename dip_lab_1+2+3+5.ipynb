Template Matching

````python
path2 = '/content/istockphoto-1219371111-612x612 (1).jpg'
path1 = '/content/istockphoto-1219371111-612x612.jpg'

img1 = cv2.imread(path1)
img2 = cv2.imread(path2)
h,w,c = img1.shape

img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
img2c = img2.copy()
result = cv2.matchTemplate(img2c, img1,cv2.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
print(min_loc, max_loc)
topleft = max_loc
bottomright = [topleft[0]+w, topleft[1]+h]
cv2.rectangle(img2c,topleft,bottomright,100,5)


fig = plt.figure(figsize=(25,25)) 
gs = GridSpec(1,3) 

fig.add_subplot(gs[0,0]) 
plt.title('Original Image') 
plt.imshow(img2, cmap='gray') 

fig.add_subplot(gs[0,1]) 
plt.title('Targeted Template') 
plt.imshow(img2c,cmap='gray')
plt.show()


#crop ball img to solve
````
